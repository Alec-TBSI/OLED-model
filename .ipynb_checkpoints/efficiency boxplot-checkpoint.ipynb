# Device parameters based on efficiency

### 1. Devices more than 50 cd/a

```python
# import the libraries we're using
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df50 = pd.read_excel('C:\Anaconda3\projects\oled\device_50.xlsx')
df50 = df50.select_dtypes(include = ['float64', 'int64']).iloc[:, 1:]
df50.head()
```

```python
#box plot
data50 = df50.drop(['thick_HIL', 'thickness_HTL1', 'thickness_HTL2', 'thickness_HTL3', 'total_HTL_thickness', 'thickness_EML', 
               'total ETL', 'thickness_ETL2', 'thickness_ETL1','thickness_EIL', 'cd/a', 'dopant_percentage'], axis=1)

data50.describe()
```

```python
f, ax = plt.subplots(figsize=(12, 10))
fig = sns.boxplot(data=data50)
plt.xticks(rotation=90,fontsize=12);
```

# Comparison of each layer

## Each layer is compared with the predicted efficiency based on the algorithm

50 means more than 50cd/a, 45 means 45-50 cd/a and so on. 5 means efficiency lower than 10

```python
HTL = {'50': df50['lumo_HTL1'] , '45': df45['lumo_HTL1'] , '40': df40['lumo_HTL1'] , '35': df35['lumo_HTL1'] , 
       '30': df30['lumo_HTL1'] , '25': df25['lumo_HTL1'] , '20': df20['lumo_HTL1'] , '10': df10['lumo_HTL1'] , 
       '5': df5['lumo_HTL1']}
sequence = ['50', '45', '40', '35', '30', '25', '20', '10', '5']
df = pd.DataFrame(data=HTL).reindex(columns=sequence)

# boxplot
f, ax = plt.subplots(figsize=(12, 10))
fig = sns.boxplot(data=df)
plt.xticks(rotation=90,fontsize=12)
plt.title('HTL LUMO ',fontsize=20);
df.describe()
```
